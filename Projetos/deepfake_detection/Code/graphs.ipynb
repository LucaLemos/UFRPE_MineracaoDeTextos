```python
import matplotlib.pyplot as plt
import numpy as np

# Performance results
models = [
    "Model CNN", 
    "Model CNN Sobel", 
    "Model CNN Sobel EfficientNet", 
    "Model CNN RNN Sobel", 
    "Model CNN Sobel LDA", 
    "Model CNN Sobel EfficientNet LDA"
]

accuracies = [0.4434, 0.9142, 0.9041, 0.8714, 0.8176, 0.4780]
roc_aucs = [0.7193, 0.9342, 0.9763, 0.8087, 0.7763, 0.9126]

# Plotting
fig, ax1 = plt.subplots(figsize=(15, 11))

# Bar plot for accuracies
ax1.bar(models, accuracies, color='b', alpha=0.6, label='Accuracy')
ax1.set_xlabel('Models')
ax1.set_ylabel('Accuracy', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Line plot for ROC-AUC
ax2 = ax1.twinx()
ax2.plot(models, roc_aucs, color='r', marker='o', label='ROC-AUC')
ax2.set_ylabel('ROC-AUC', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Titles and legend
plt.title('Performance Comparison of Different Models')
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
models = [
    "CNN", "CNN Sobel", "CNN Sobel EfficientNet", 
    "CNN RNN Sobel", "CNN Sobel LDA", "CNN Sobel EfficientNet LDA"
]
accuracies = [0.4434, 0.9142, 0.9041, 0.8714, 0.8176, 0.4780]
roc_aucs = [0.7193, 0.9342, 0.9763, 0.8087, 0.7763, 0.9126]

# Plotting individual graphs for each model
for i, model in enumerate(models):
    plt.figure(figsize=(8, 6))
    plt.bar(["Accuracy", "ROC-AUC"], [accuracies[i], roc_aucs[i]], color=["blue", "green"])
    plt.ylim(0, 1.1)
    plt.title(f"Performance of {model}")
    plt.ylabel("Score")
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(models, accuracies, marker='o', label='Accuracy', color='blue')
plt.plot(models, roc_aucs, marker='o', label='ROC-AUC', color='green')
plt.ylim(0, 1.1)
plt.title("Model Performance Comparison")
plt.xlabel("Models")
plt.ylabel("Score")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
